````python
import glob
import pandas as pd
````

````python
target_files = sorted(glob.glob("project-docs/*txt"))
target_files
````

````python
for target_file in target_files:
    with open(target_file, "r") as f:
        all_text = f.readlines()
        print(f"Year: {all_text[0].strip()}")
        print(f"Medical Condition: {all_text[2].strip()}")
        print(f"Care Hours: {all_text[4].strip()}")
            
````

````python
data = []
for target_file in target_files:
    with open(target_file, "r") as f:
        all_text = f.readlines()
        year = all_text[0].strip()
        print(year)
        medical_condition = all_text[2].strip()
        print(medical_condition)
        care_hours = all_text[4].strip()
        print(care_hours)
        if "Dementia" in medical_condition:
                    cognition_related = True
                    medical_condition = 'Dementia'
        elif "Alzheimer" in medical_condition:
                    cognition_related = True
                    medical_condition = 'Alzheimer’s'
        else:
            cognition_related = False
            medical_condition = 'Not Specified'
        print(cognition_related)
        data.append({
                    'year': year,
                    'cognition_related': cognition_related,
                    'medical_condition': medical_condition,
                    'care_hours': care_hours
                })

        print("*******")
````

````python
df = pd.DataFrame(data)
df
````
